Select features most relevant to crisis prediction and treatment recommendation.

```python
# Select relevant features
features = ['WBC', 'HGB', 'LYMp', 'NEUTp'] 
cbc_data = cbc_data[features]
print("\nSelected Features:")
print(cbc_data.head())
```

Define crisis_status - which is meant to determine when a patient is most likely to have a crisis or not

```python
# Define crisis status based on medical thresholds
cbc_data['crisis_status'] = np.where((cbc_data['HGB'] < 9) & (cbc_data['WBC'] > 13), 1, 0)

print("\nCrisis Status Distribution:")
print(cbc_data['crisis_status'].value_counts())
```

WBC distribution 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of WBC values
plt.figure(figsize=(10, 6))
sns.histplot(cbc_data['WBC'], kde=True, bins=30, color='blue')
plt.title('Distribution of WBC Values')
plt.xlabel('WBC (x10/L)')
plt.ylabel('Frequency')
plt.show()
```

Now we do the preprocessing of data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


X = cbc_data[['WBC', 'HGB', 'LYMp', 'NEUTp']]  
y_crisis = cbc_data['crisis_status']


X_train, X_test, y_train, y_test = train_test_split(X, y_crisis, test_size=0.2, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

This shows the distribution of the crisis status in both the training and testing sets

```python
# Check class distribution before splitting
print("Class distribution (overall):")
print(cbc_data['crisis_status'].value_counts())

# Check class distribution in the training set
print("\nClass distribution (training set):")
print(y_train.value_counts())

# Check class distribution in the test set
print("\nClass distribution (test set):")
print(y_test.value_counts())
```

We now train the crisis detection model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# Train Logistic Regression model
crisis_model = LogisticRegression(class_weight='balanced', random_state=42)
crisis_model.fit(X_train_scaled, y_train)

# Predictions
y_pred = crisis_model.predict(X_test_scaled)
y_pred_prob = crisis_model.predict_proba(X_test_scaled)[:, 1]

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

print(f"Accuracy: {accuracy}")

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
train_accuracy = crisis_model.score(X_train_scaled, y_train)
test_accuracy = crisis_model.score(X_test_scaled, y_test)
print(f"Training Accuracy: {train_accuracy}")
print(f"Testing Accuracy: {test_accuracy}")
```

Here we look at the confusion matrix and the roc curve for evaluation

```python
# Confusion Matrix Visualization
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Crisis', 'Crisis'], yticklabels=['No Crisis', 'Crisis'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ROC Curve Visualization
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

We then train the treatment recommendation model

```python
from sklearn.ensemble import RandomForestClassifier

# Define treatment classes 
cbc_data['treatment_class'] = np.where((cbc_data['HGB'] < 9) & (cbc_data['WBC'] > 13), 'severe-pharma', 'non-pharma')

# Features for treatment model
X_treatment = cbc_data[['WBC', 'HGB', 'crisis_status']]
y_treatment = cbc_data['treatment_class']

# Split into train-test
X_train_treat, X_test_treat, y_train_treat, y_test_treat = train_test_split(X_treatment, y_treatment, test_size=0.2, random_state=42)

# Train Random Forest
treatment_model = RandomForestClassifier(class_weight='balanced', random_state=42)
treatment_model.fit(X_train_treat, y_train_treat)

# Predictions and accuracy
y_pred_treat = treatment_model.predict(X_test_treat)
accuracy_treat = accuracy_score(y_test_treat, y_pred_treat)

print(f"Treatment Recommendation Accuracy: {accuracy_treat}")
```

This shows the level of importance of the features used to determine treatment

```python
# Feature importance for treatment recommendation
feature_importances = pd.Series(treatment_model.feature_importances_, index=X_treatment.columns)
feature_importances.sort_values(ascending=False).plot(kind='bar', color='red',title='Feature Importance', figsize=(8, 6))
plt.show()
```
